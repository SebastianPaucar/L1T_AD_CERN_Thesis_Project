```python
def convert_to_h5(input_file, output_file, prescale_file, tree_name, uGT_tree_name, event_tree_name, gen_tree_name, mc, hw):
    inFile = uproot.open(input_file)
    l1Tree = inFile[tree_name]
    uGTTree = inFile[uGT_tree_name]
    evtTree = inFile[event_tree_name]
    nentries = l1Tree.num_entries

    bit_arrays = uGTTree.arrays(['m_algoDecisionInitial', 'm_algoDecisionFinal'], library='np')
    initial_bits = np.stack(bit_arrays['m_algoDecisionInitial'], axis=0)
    final_bits = np.stack(bit_arrays['m_algoDecisionFinal'], axis=0)

    algo_map = getAlgoMap(input_file, uGT_tree_name)
    algo_map = filterAlgoMap(algo_map,prescale_file)

    seeds = {seedname: np.empty([nentries], dtype=bool) for seedname in algo_map.keys()}
    for seedname, bit in algo_map.items():
        seeds[seedname][:] = final_bits[:,bit].astype(bool)
    seeds["L1bit"] = np.logical_or.reduce([seeds[seedname] for seedname in algo_map.keys()]).astype(bool)

    #njets = 10
    #nmuons = 4
    #nelectrons = 4
    njets = 12
    nmuons = 8
    nelectrons = 12
    cylNames = ['pT', 'eta', 'phi']
    # variables to retrieve
    evtInfoList = ["run","lumi","event","bx","orbit","time","nPV_True"]
    if hw == True:
        varList = ['nSums', 'sumType', 'sumIEt', 'sumIPhi', 'sumBx',
                   'jetIEt', 'jetRawEt', 'jetIEta', 'jetIPhi', 'jetBx',
                   'muonIEt', 'muonIEta', 'muonIPhi', 'muonIso', 'muonDxy', 'muonIEtUnconstrained', 'muonQual', 'muonBx',
                   'egIEt', 'egIEta', 'egIPhi', 'egIso', 'egBx']
    else:
        varList = ['nSums', 'sumType', 'sumEt', 'sumPhi', 'sumBx','jetBx','muonBx','egBx',
                   'jetEt', 'jetRawEt', 'jetEta', 'jetPhi', 'jetBx',
                   'muonEt', 'muonEta', 'muonPhi', 'muonIso', 'muonDxy', 'muonEtUnconstrained', 'muonQual', 'muonBx',
                   'egEt', 'egEta', 'egPhi', 'egIso']                       
       
    # get awkward arrays
    arrays = l1Tree.arrays(varList)
    evtArray = evtTree.arrays(evtInfoList)
    
    # sums: store the following
    # kTotalEt, kTotalEtEm, kTotalHt, kMissingEt, kMissingHt,
    # with type 0, 16, 1, 2, 3
    l1sum_cyl = np.zeros((nentries, 3), dtype=np.float16)
    sumType = to_np_array(arrays['sumType'], maxN=arrays['nSums'][0])
    sumBx = to_np_array(arrays['sumBx'], maxN=arrays['nSums'][0])
    if hw == True:
        sumEt = to_np_array(arrays['sumIEt'], maxN=arrays['nSums'][0])
        sumPhi = to_np_array(arrays['sumIPhi'], maxN=arrays['nSums'][0])
    else:
        sumEt = to_np_array(arrays['sumEt'], maxN=arrays['nSums'][0])
        sumPhi = to_np_array(arrays['sumPhi'], maxN=arrays['nSums'][0])     
    # index of sum to save (MET)
    metindex = np.where( (sumType == 2) & (sumBx == 0) )
    l1sum_cyl[:,0] = sumEt[metindex] # MET_pt
    l1sum_cyl[:,2] = sumPhi[metindex] # MET_phi

    l1ht = np.zeros((nentries,), dtype=np.float16)
    # index of sum to save HT
    metindex = np.where( (sumType == 1) & (sumBx == 0) )
    l1ht[:] = sumEt[metindex]

    l1et = np.zeros((nentries,), dtype=np.float16)
    # index of sum to save ET
    metindex = np.where( (sumType == 0) & (sumBx == 0) )
    l1et[:] = sumEt[metindex]
        
    # store objects: jets, muons, electrons
    if hw == True:
        l1Jet_cyl, _, _, _, _ = store_objects(arrays, nentries, nobj=njets, obj='jetI')
        l1mu_cyl, l1mu_iso, l1mu_dxy, l1mu_upt, l1mu_qual = store_objects(arrays, nentries, nobj=nmuons, obj='muonI')
        l1ele_cyl, l1ele_iso, _, _, _ = store_objects(arrays, nentries, nobj=nelectrons, obj='egI')
    else:
        l1Jet_cyl, _, _, _, _ = store_objects(arrays, nentries, nobj=njets, obj='jet')
        l1mu_cyl, l1mu_iso, l1mu_dxy, l1mu_upt = store_objects(arrays, nentries, nobj=nmuons, obj='muon')
        l1ele_cyl, l1ele_iso, _, _, _ = store_objects(arrays, nentries, nobj=nelectrons, obj='eg')

    genHT = np.zeros((nentries,), dtype=np.float16)
    if mc == True:
        
        genTree = inFile[gen_tree_name]
        genInfoList = ["jetPt","jetEta"]
        genArray = genTree.arrays(genInfoList)
        genJets = np.zeros((nentries, 100, 2), dtype=np.float16)
        pt = to_np_array(genArray['jetPt'], maxN=100)
        eta = to_np_array(genArray['jetEta'], maxN=100)
        genJets[:,:,0] = pt
        genJets[:,:,1] = eta
        mask0 = genJets[:,:,0] > 30
        mask1 = genJets[:,:,1] < 2.5
        mask = np.concatenate((mask0[:,:,np.newaxis],mask1[:,:,np.newaxis]),axis=2)*1
        genJets = genJets*mask
        genHT = np.sum(genJets[:,:,0],axis=1)

    outFile = h5py.File(output_file, 'w')
    outFile.create_dataset('FeatureNames_cyl', data=cylNames, compression='gzip')
    outFile.create_dataset('l1Jet_cyl', data=l1Jet_cyl, compression='gzip')
    outFile.create_dataset('l1Muon_cyl', data=l1mu_cyl, compression='gzip')
    outFile.create_dataset('l1Muon_Iso', data=l1mu_iso, compression='gzip')
    outFile.create_dataset('l1Muon_Qual', data=l1mu_qual, compression='gzip')
    outFile.create_dataset('l1Muon_Dxy', data=l1mu_dxy, compression='gzip')
    outFile.create_dataset('l1Muon_Upt', data=l1mu_upt, compression='gzip')
    outFile.create_dataset('l1Ele_cyl', data=l1ele_cyl, compression='gzip')
    outFile.create_dataset('l1Ele_Iso', data=l1ele_iso, compression='gzip')
    outFile.create_dataset('l1Sum_cyl', data=l1sum_cyl, compression='gzip')
    outFile.create_dataset('l1ET', data=l1et, compression='gzip')
    outFile.create_dataset('l1HT', data=l1ht, compression='gzip')
    outFile.create_dataset('EventInfoNames', data=evtInfoList, compression='gzip')
    outFile.create_dataset('EventInfo', data=ak.to_pandas(evtArray).to_numpy(), compression='gzip')
    outFile.create_dataset('genHT', data=genHT, compression='gzip')
    for seed, values in seeds.items():
        outFile.create_dataset(seed, data=values, compression='gzip')
    outFile.close()
```

```python
def convert_to_h5(input_file, output_file, prescale_file, tree_name, uGT_tree_name, event_tree_name, gen_tree_name, mc, hw):
    inFile = uproot.open(input_file)
    l1Tree = inFile[tree_name]
    uGTTree = inFile[uGT_tree_name]
    evtTree = inFile[event_tree_name]
    nentries = l1Tree.num_entries

    bit_arrays = uGTTree.arrays(['m_algoDecisionInitial', 'm_algoDecisionFinal'], library='np')
    initial_bits = np.stack(bit_arrays['m_algoDecisionInitial'], axis=0)
    final_bits = np.stack(bit_arrays['m_algoDecisionFinal'], axis=0)

    algo_map = getAlgoMap(input_file, uGT_tree_name)
    algo_map = filterAlgoMap(algo_map,prescale_file)

    seeds = {seedname: np.empty([nentries], dtype=bool) for seedname in algo_map.keys()}
    for seedname, bit in algo_map.items():
        seeds[seedname][:] = final_bits[:,bit].astype(bool)
    seeds["L1bit"] = np.logical_or.reduce([seeds[seedname] for seedname in algo_map.keys()]).astype(bool)

    #njets = 10
    #nmuons = 4
    #nelectrons = 4
    njets = 12
    nmuons = 8
    nelectrons = 12
    ntaus = 12
    cylNames = ['pT', 'eta', 'phi']
    # variables to retrieve
    evtInfoList = ["run","lumi","event","bx","orbit","time","nPV_True"]
    if hw == True:
        varList = ['nSums', 'sumType', 'sumIEt', 'sumIPhi', 'sumBx',
                   'jetIEt', 'jetRawEt', 'jetIEta', 'jetIPhi', 'jetBx', 'jetHwQual', 
                   'muonIEt', 'muonIEtaAtVtx', 'muonIPhiAtVtx', 'muonIso', 'muonDxy', 'muonIEtUnconstrained', 'muonQual', 'muonBx',
                   'egIEt', 'egIEta', 'egIPhi', 'egIso', 'egBx',
                   'tauIEt', 'tauIEta', 'tauIPhi', 'tauIso', 'tauBx']
    else:
        varList = ['nSums', 'sumType', 'sumEt', 'sumPhi', 'sumBx','jetBx','muonBx','egBx',
                   'jetEt', 'jetRawEt', 'jetEta', 'jetPhi', 'jetBx', 'jetHwQual', 
                   'muonEt', 'muonEtaAtVtx', 'muonPhiAtVtx', 'muonIso', 'muonDxy', 'muonEtUnconstrained', 'muonQual', 'muonBx',
                   'egEt', 'egEta', 'egPhi', 'egIso',
                   'tauEt', 'tauEta', 'tauPhi', 'tauIso', 'tauBx']                    
       
    # get awkward arrays
    arrays = l1Tree.arrays(varList)
    evtArray = evtTree.arrays(evtInfoList)
    
    # sums: store the following
    # kTotalEt, kTotalEtEm, kTotalHt, kMissingEt, kMissingHt,
    # with type 0, 16, 1, 2, 3
    l1sum_cyl = np.zeros((nentries, 3), dtype=np.float16)
    sumType = to_np_array(arrays['sumType'], maxN=arrays['nSums'][0])
    sumBx = to_np_array(arrays['sumBx'], maxN=arrays['nSums'][0])
    if hw == True:
        sumEt = to_np_array(arrays['sumIEt'], maxN=arrays['nSums'][0])
        sumPhi = to_np_array(arrays['sumIPhi'], maxN=arrays['nSums'][0])
    else:
        sumEt = to_np_array(arrays['sumEt'], maxN=arrays['nSums'][0])
        sumPhi = to_np_array(arrays['sumPhi'], maxN=arrays['nSums'][0])     
    # index of sum to save (MET)
    metindex = np.where( (sumType == 2) & (sumBx == 0) )
    l1sum_cyl[:,0] = sumEt[metindex] # MET_pt
    l1sum_cyl[:,2] = sumPhi[metindex] # MET_phi

    l1ht = np.zeros((nentries,), dtype=np.float16)
    # index of sum to save HT
    metindex = np.where( (sumType == 1) & (sumBx == 0) )
    l1ht[:] = sumEt[metindex]

    l1et = np.zeros((nentries,), dtype=np.float16)
    # index of sum to save ET
    metindex = np.where( (sumType == 0) & (sumBx == 0) )
    l1et[:] = sumEt[metindex]
        
    # store objects: jets, muons, electrons
    if hw == True:
        l1Jet_cyl, _, _, _, l1Jet_qual = store_objects(arrays, nentries, nobj=njets, obj='jetI')
        l1mu_cyl, l1mu_iso, l1mu_dxy, l1mu_upt, l1mu_qual = store_objects(arrays, nentries, nobj=nmuons, obj='muonI')
        l1ele_cyl, l1ele_iso, _, _, _ = store_objects(arrays, nentries, nobj=nelectrons, obj='egI')
        l1tau_cyl, l1tau_iso, _, _, _ = store_objects(arrays, nentries, nobj=ntaus, obj='tauI')
    else:
        l1Jet_cyl, _, _, _, _ = store_objects(arrays, nentries, nobj=njets, obj='jet')
        l1mu_cyl, l1mu_iso, l1mu_dxy, l1mu_upt = store_objects(arrays, nentries, nobj=nmuons, obj='muon')
        l1ele_cyl, l1ele_iso, _, _, _ = store_objects(arrays, nentries, nobj=nelectrons, obj='eg')
        l1tau_cyl, l1tau_iso, _, _, _ = store_objects(arrays, nentries, nobj=ntaus, obj='tau')

    genHT = np.zeros((nentries,), dtype=np.float16)
    if mc == True:
        
        genTree = inFile[gen_tree_name]
        genInfoList = ["jetPt","jetEta"]
        genArray = genTree.arrays(genInfoList)
        genJets = np.zeros((nentries, 100, 2), dtype=np.float16)
        pt = to_np_array(genArray['jetPt'], maxN=100)
        eta = to_np_array(genArray['jetEta'], maxN=100)
        genJets[:,:,0] = pt
        genJets[:,:,1] = eta
        mask0 = genJets[:,:,0] > 30
        mask1 = genJets[:,:,1] < 2.5
        mask = np.concatenate((mask0[:,:,np.newaxis],mask1[:,:,np.newaxis]),axis=2)*1
        genJets = genJets*mask
        genHT = np.sum(genJets[:,:,0],axis=1)

    outFile = h5py.File(output_file, 'w')
    outFile.create_dataset('FeatureNames_cyl', data=cylNames, compression='gzip')
    outFile.create_dataset('l1Jet_cyl', data=l1Jet_cyl, compression='gzip')
    outFile.create_dataset('l1Jet_Qual', data=l1Jet_qual, compression='gzip') 
    outFile.create_dataset('l1Muon_cyl', data=l1mu_cyl, compression='gzip')
    outFile.create_dataset('l1Muon_Iso', data=l1mu_iso, compression='gzip')
    outFile.create_dataset('l1Muon_Qual', data=l1mu_qual, compression='gzip')
    outFile.create_dataset('l1Muon_Dxy', data=l1mu_dxy, compression='gzip')
    outFile.create_dataset('l1Muon_Upt', data=l1mu_upt, compression='gzip')
    outFile.create_dataset('l1Ele_cyl', data=l1ele_cyl, compression='gzip')
    outFile.create_dataset('l1Ele_Iso', data=l1ele_iso, compression='gzip')
    outFile.create_dataset('l1Tau_cyl', data=l1tau_cyl, compression='gzip')
    outFile.create_dataset('l1Tau_Iso', data=l1tau_iso, compression='gzip')    
    outFile.create_dataset('l1Sum_cyl', data=l1sum_cyl, compression='gzip')
    outFile.create_dataset('l1ET', data=l1et, compression='gzip')
    outFile.create_dataset('l1HT', data=l1ht, compression='gzip')
    outFile.create_dataset('EventInfoNames', data=evtInfoList, compression='gzip')
    outFile.create_dataset('EventInfo', data=ak.to_pandas(evtArray).to_numpy(), compression='gzip')
    outFile.create_dataset('genHT', data=genHT, compression='gzip')
    for seed, values in seeds.items():
        outFile.create_dataset(seed, data=values, compression='gzip')
    outFile.close()
```
